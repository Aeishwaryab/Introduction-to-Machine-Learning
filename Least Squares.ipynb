# Least Square

```python
# least Square
X = np.linspace(0,1, num=1000)[np.newaxis]
X = np.transpose(X)
X.shape
```

```python
mu ,var = 0, 0.25
sig = np.random.normal(mu,var,1000)[np.newaxis]
sig = sig.T
np.disp(sig.size) 
sig.shape
```

```python
y = X+ sig
y.size
```

```python
plt.scatter(X,y)
plt.show()
```

```python
X.shape
```

```python
Xt = np.transpose(X)
Xt.shape
```

```python
XtX = np.matmul(Xt,X)
XtX.shape 
```

```python
Xty = np.matmul(Xt, y)
```

```python
a = Xty/XtX
print(a)
```

```python
gx = a*X
print(gx.shape)
```

```python
plt.scatter(X,y)
plt.plot(X,gx,'r','-o')
plt.show()
```

# Question 4.4
## input matrix and error

```python
X = np.random.uniform(0,1,1000)[np.newaxis]
X = X.T
mu ,var = 0, 0.1
sig = np.random.normal(mu,var,1000)[np.newaxis]
sig = sig.T
```

## output

```python
X_025 = X-0.25
X_012 = np.square(X_025)
X_075 = X-0.75
X_022 = np.square(X_075)
X_01 = X_012*X_022
Y = 30*X_01 + sig
Y.shape
```

## plot

```python
plt.scatter(X,Y)
plt.show()
```

```python
shp = X.shape
print(shp)
```

```python
P = np.concatenate((np.ones(shp), X, np.square(X), X*np.square(X), np.square(np.square(X))),axis=1)
```

```python
Pt = P.T
PtP = np.matmul(Pt,P)
PtP_inv = np.linalg.inv(PtP)
PtY = np.matmul(Pt, Y)
A = np.matmul(PtP_inv,PtY)
X.shape
```

```python
print(A)
gx = np.matmul(P,A)
gx.shape
```

```python
plt.scatter(X,Y)
plt.plot(X,gx,'ro')
plt.show()
```
